# Aggregate EPL

```python
import os 
from statsbombpy import sb 
import warnings 
#ignore by message 
warnings.filterwarnings("ignore", message="credentials were not supplied")

import pandas as pd 
from tqdm import tqdm
```

# Competition 

```python
df_competitions= sb.competitions()
```

# Extract EPL

```python
df_epl= df_competitions[ df_competitions.competition_name == "Premier League" ] 
df_epl
```

```python
for season_id in df_epl.season_id:
    print(season_id)
```

# Get Events

```python
all_events = []

for season_id in df_epl.season_id:

    # get match 
    df_match= sb.matches(competition_id=2, season_id=season_id)

    # get events
    for match_id in tqdm(df_match.match_id, desc=f"{season_id}"):
        df_events= sb.events(match_id= match_id)
        all_events.append(df_events)
```

```python
df_all_events=pd.concat(all_events, ignore_index=True)
```

```python
del all_events
```

```python
df_all_events
```

# Extract Only Shots

```python
df_all_shots = df_all_events[ (df_all_events.type== "Shot") ]
df_all_shots
```

# Write to File

```python
path_data= "./data"
```

```python
os.makedirs(path_data, exist_ok= True)
```

```python
df_all_events.to_parquet(
    os.path.join(path_data, "data-epl-all.parquet")
)
```

```python
df_all_shots.to_parquet(
    os.path.join(path_data, "data-epl-shots.parquet")
)
```
