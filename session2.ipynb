# shot

## Position

```python
# shot position
x= 108.3 
y= 44.6 
```

```python
pitch= Pitch(
    pitch_type="statsbomb", 
    pitch_color= "grass",
    line_color= "white", 
    stripe= True, 
    label= True
)
fig, ax= pitch.draw()
pitch.scatter(x, y, ax=ax, color= "yellow") 
plt.show()
```

## Player Name 

```python
name= "Son"
```

```python
pitch= Pitch( 
    pitch_type="statsbomb", 
    pitch_color= "grass", 
    line_color= "white", 
    stripe= True
)
fig, ax= pitch.draw()
pitch.scatter(x, y, ax=ax, color="yellow")
pitch.annotate(name, (x-5, y+5), ax=ax, color="white")
plt.show()
```

## marker, color, and size

```python
pitch= Pitch(
    pitch_type="statsbomb", 
    pitch_color="grass", 
    line_color="white", 
    stripe=True
)
fig, ax= pitch.draw()

pitch.scatter(30, 20, ax=ax, color= "black", marker="x", s= 50)
pitch.scatter(50, 20, ax=ax, color="yellow", marker="s", s=250) 
pitch.scatter(70, 20, ax=ax, color="silver", marker="*", s=450)
pitch.scatter(90, 20, ax=ax, marker="football", s= 650)

plt.show()
```

# Pass

## Basic

```python
x1= 78
y1= 40
x2= 29 
y2= 30 
```

```python
pitch= Pitch(
    pitch_type= "statsbomb", 
    pitch_color= "grass", 
    line_color= "white",
    stripe= True
) 
fig, ax= pitch.draw()

pitch.lines(x1, y1, x2, y2, ax=ax, color="gold") 
plt.show()
```

## comet

```python
pitch= Pitch( 
    pitch_type="statsbomb", 
    pitch_color="grass", 
    line_color= "white", 
    stripe= True
) 
fig, ax= pitch.draw() 
pitch.lines(x1, y1, x2, y2, ax=ax, color="gold", comet= True) 
plt.show()
```

## without Z-Order

```python
pitch= Pitch( 
    pitch_type="statsbomb", 
    pitch_color= "grass", 
    line_color= "white", 
    stripe= True
) 
fig, ax = pitch.draw() 

pitch.scatter(x2, y2, ax=ax, marker= "football", s= 300) 
pitch.lines(x1, y1, x2, y2, ax=ax, color="gold", comet=True) 

plt.show()
```

## with Z-Order

```python
pitch= Pitch( 
    pitch_type="statsbomb", 
    pitch_color="grass", 
    line_color= "white", 
    stripe= True
) 
fig, ax = pitch.draw()

pitch.scatter(x2, y2, ax=ax, marker="football", s= 300, zorder=999)
pitch.lines(x1, y1, x2, y2, ax=ax, color= "gold", comet= True) 

plt.show()
```

## Arrow

```python
pitch= Pitch( 
    pitch_color="grass", 
    line_color="white",
    stripe= True
)
fig, ax= pitch.draw()

pitch.arrows(x1, y1, x2, y2, ax=ax, color="gold")
plt.show()
```

## Line Width 

```python
pitch= Pitch(
    pitch_color="grass",
    line_color="white", 
    stripe= True
) 
fig, ax = pitch.draw()
pitch.arrows(x1, y1, x2, y2, ax=ax, color="gold", width=20) 
plt.show()
```

## Transparency 

```python
pitch= Pitch( 
    pitch_color="grass", 
    line_color="white", 
    stripe= True
) 
fig, ax= pitch.draw() 

pitch.arrows(x1, y1, x2, y2, ax=ax, color="gold", width=20, alpha=0.7) 
plt.show()
```
