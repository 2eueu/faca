## Visualization 

```python
import matplotlib.pyplot as plt 
from mplsoccer import Pitch
```

```python
df_yellow_cards = pd.DataFrame({
    'location': [[6.4, 62.0], [72.3, 21.1], [72.1, 65.8], [47.2, 12.1]],  # 위치 좌표
    'player': ['Héctor Bellerín Moruno', 'Erik Lamela', 'Francis Joseph Coquelin', 'Eric Dier']
})


pitch = Pitch(
    pitch_type="statsbomb",
    pitch_color="grass",
    line_color="white",
    stripe=True
)

fig, ax = pitch.draw()

# 데이터 처리
df = df_yellow_cards.squeeze()

# 각 위치 데이터를 순회하며 처리
for index, row in df.iterrows():
    # location 데이터를 x, y로 분리
    location = row['location']
    x, y = location  # 리스트에서 좌표 분리

    # 위치에 따라 스캐터 플롯
    pitch.scatter(x, y, color="yellow", s=100, ax=ax)

    # 선수 이름을 플롯에 추가
    pitch.annotate(
        row['player'],  # 선수 이름
        (x - 10, y - 3),  # 위치 조정
        color="white", fontweight="bold",
        ax=ax
    )

plt.show()
```
